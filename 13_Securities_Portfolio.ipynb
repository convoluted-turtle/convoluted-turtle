<h1><center>Stock returns on select S&P 500 Companies</center></h1>

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['MMM', 'AAL', 'SNY', 'UNH', 'WMT', 'GOOG', 'AMZN', 'MSFT', 'BA', 'V', 'GS', 'JPM', 'DIS']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader (t, data_source = 'yahoo', start = '2009-1-1')['Adj Close']
```

```python
mydata.info()
```

```python
mydata.iloc[2490]
# Extracts Data from Last row of the table. Last price when this file was created: 11/21/2018
```

```python
(mydata / mydata.iloc[0] * 100).plot(figsize = (15,6))
# Normalizing the data (pt/p0) * 100
```

# Calculating Returns

```python
returns = (mydata / mydata.shift(1)) - 1
# simple returns approach instead of log because it is preferred approach for time series analysis on street
```

```python
weights = np.array([0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,0.0769,])
```

```python
np.dot(returns, weights)
```

```python
annual_returns = returns.mean() * 253
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
portfolio_1 = str(round(np.dot(annual_returns, weights) , 5) * 100) + '%'
print(portfolio_1)
# Annualized returns of equally weighted portfolio in 13 stocks
```
